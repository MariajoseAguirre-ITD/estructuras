```python
def DigitoDeLaVida(x):
    sum = 0
    
    while x:
        sum += x % 10
        x //= 10
    
    if sum > 9:
        return DigitoDeLaVida(sum)

    return sum

x = int(input("Inserta la fecha de tu cumplea単os (en digitos): "))

print(DigitoDeLaVida(x))
```
